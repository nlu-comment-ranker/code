```python
from sqlalchemy import *
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import relation, sessionmaker
```

```python
import commentDB
reload(commentDB)

engine = create_engine('sqlite:///redditDB.sqlite')

Session = sessionmaker(bind=engine)
session = Session()
```

```python
com_query = session.query(commentDB.Comment)
print com_query.count()

# Unfortunately, this query is too sophisticated
# and we can't run it directly using query.filter
coms = [c for c in com_query if c.user and c.user.user_id and len(c.user.activities) > 0]
print len(coms)
```

```python
import features
reload(features)

import time

featureList = [features.FeatureSet(c, user=c.user) for c in coms[:100]]

t0 = time.time()
t1 = time.time()
counter = 0
for f in featureList:
    f.tokenize()
    f.calc_token_counts()
    f.calc_SMOG()
    
    f.pos_tag()
    f.calc_nouns_verbs()
    
    f.stem()
    f.calc_entropy()
    
    f.calc_user_activity_features()
    
    # Call this to recover memory!
    f.clean_temp()
    
    counter += 1
    if counter % 100 == 0:
        temp = t1
        t1 = time.time()
        print "Last 100: %02f s" % (t1 - temp)

print "Completed %d in %.02f s" % (counter, time.time() - t0)
```
