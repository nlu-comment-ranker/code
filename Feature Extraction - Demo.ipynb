## Extract Features from Comment List

```python
import features
reload(features)

import nltk
import time, sys

nsamples = 1000
featureSets = [features.FeatureSet(c, user=c.user, parent=c.submission) for c in coms[:nsamples]]
printevery = 1000
```

```python
vsm = features.VSM(featureSets)
vsm.index_featuresets(tag="_global")
vsm.build_VSM(lowercase=True, max_df=1.0)
vsm.build_TFIDF()
```

```python
t0 = time.time()
t1 = time.time()
counter = 0
for f in featureSets:
    f.tokenize()
    f.calc_token_counts()
    f.calc_SMOG()
    
    # These are slow! Enable only if needed.
#     f.pos_tag()
#     f.calc_nouns_verbs()
    
    f.calc_entropy(vsmTag="_global")
    
    # This won't run unless user data is available
#     f.calc_user_activity_features()

    f.calc_parent_rank_features()
    
    f.calc_parent_overlap(vsmTag="_global")
    
    # Call this to recover memory!
    f.clean_temp()
    
    counter += 1
    if counter % printevery == 0:
        temp = t1
        t1 = time.time()
        print >> sys.stderr, "Last %d: %02f s" % (printevery, (t1 - temp))

print >> sys.stderr, "Completed %d in %.02f s" % (counter, time.time() - t0)
```
